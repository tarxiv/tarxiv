```python
# Now lets try a search (redshift < 0.025, peak_mag between 14.5, 16.0, in V or g filter, type SN)
search = {
    "redshift": [{
        "value": 0.025,
        "operator": "<"}],
    "peak_mag": [
        {"value": 14.5,
         "operator": ">"},
        {"filter": ["g", "v"],
         "operator": "IN"}],
    "object_type": [
        {"value": "SN",
         "operator": "="}]
}
# Test a simple object pull
route = "/search_objects"
# Run query
response = requests.post(url=url + route, json={"search": search}, headers={'Authorization': 'TOKEN'})
response.json()
```
